```python
from matplotlib import pyplot as plt
import matplotlib.image as mpimg
import numpy as np
```

```python
img = mpimg.imread('test.jpg')
```

```python
R,G,B =img[:,:,0], img[:,:,1], img[:,:,2]
```

```python
imgGray = 0.2989*R+0.5870*G+0.1140*B
```

```python
plt.imshow(imgGray,cmap='gray')
plt.show()
```

```python
# Tamanho no eixo 0
np.size(imgGray,0)
```

```python
# Tamanho no eixo 1
np.size(imgGray,1)
```

```python
imgGray[1,1]
```

```python
k=140
imgBW=imgGray.copy()
for i in range(np.size(imgGray,0)-1):
    for j in range(np.size(imgGray,1)-1):
        if imgGray[i,j]>k:
            imgBW[i,j]=255
        else:
            imgBW[i,j]=0
```

```python
plt.imshow(imgBW,cmap='gray')
plt.show()
```
